```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 1. Création des données simulées (Corridor Lomé - Sahel)
data = {
    "Prestenaire": ["SOTOTRANS", "SAHEL LOG", "AFRIQ TRANS", "NIGER ROUTE"],
    "Tarif_FCFA": [900000, 1950000, 1800000, 2400000],
    "SLA_Respect_Delais": [94, 82, 68, 88],
    "Axe_Routier": [
        "Lomé - Cinkassé",
        "Lomé - Ouagadougou",
        "Lomé - Ouagadougou",
        "Lomé - Niamey",
    ],
}
df = pd.DataFrame(data)

# 2. Configuration du graphique double axe (Tarif vs Fiabilité)
sns.set_theme(style="whitegrid")
fig, ax1 = plt.subplots(figsize=(10, 6))

# Axe 1 : Barres pour les Tarifs
sns.barplot(
    x="Prestenaire",
    y="Tarif_FCFA",
    data=df,
    ax=ax1,
    palette="Blues_r",
    alpha=0.8,
)
ax1.set_title(
    "Analyse Comparative des Prestataires : Tarifs vs Fiabilité (SLA)",
    fontsize=14,
    fontweight="bold",
    pad=15,
)
ax1.set_ylabel("Tarif Moyen (FCFA)", fontsize=12, fontweight="bold")
ax1.set_xlabel("Prestataires de Transport", fontsize=12)

# Axe 2 : Ligne pour le taux de respect des délais (SLA)
ax2 = ax1.twinx()
sns.lineplot(
    x="Prestenaire",
    y="SLA_Respect_Delais",
    data=df,
    ax=ax2,
    color="crimson",
    marker="o",
    linewidth=2.5,
    markersize=8,
)
ax2.set_ylabel(
    "Taux de Respect des Délais (%)", fontsize=12, color="crimson", fontweight="bold"
)
ax2.set_ylim(50, 100)

import os

# 1. Vérifie si le dossier existe, sinon il le crée automatiquement
dossier = "projet1_tms_optimisation"
if not os.path.exists(dossier):
    os.makedirs(dossier)

# 2. Ajustement des marges et sauvegarde de l'image
plt.tight_layout()
plt.savefig(f"{dossier}/comparatif_prestataires.png", dpi=300)
print(f"Succès ! L'image a été sauvegardée dans le dossier : {dossier}/")
plt.show()
# Ajout d'une ligne d'alerte à 80% de SLA (Seuil Critique)
ax2.axhline(80, color="orange", linestyle="--", label="Seuil Critique SLA (80%)")

```
